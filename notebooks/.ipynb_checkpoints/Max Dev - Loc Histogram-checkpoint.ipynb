```python
# import modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data into a dataframe
path = '../data/GP_Alumni_List_Cleaned.csv'
df = pd.read_csv(path).copy()
df
```

```python
# value count graph for states
value_counts = df['State'].value_counts()

fig, ax = plt.subplots(figsize=(20, 6))
ax.bar(value_counts.index, value_counts.values, color='#879EC3')
ax.set_xlabel('States')
ax.set_ylabel('# of Alumni')
ax.set_title('Alumni Distribution in the USA')
plt.savefig('../figures/Alumni_Dist_US.png')
```

```python
# value count graph for states
value_counts = df['Country'].value_counts()

fig, ax = plt.subplots(figsize=(20, 6))
ax.bar(value_counts.index, value_counts.values, color='#CC4628')
ax.set_xticklabels(value_counts.index, rotation=45, horizontalalignment='right')

ax.set_xlabel('Countries')
ax.set_ylabel('# of Alumni')
ax.set_title('Alumni Distribution Internationally')
plt.savefig('../figures/Alumni_Dist_International.png')
```
