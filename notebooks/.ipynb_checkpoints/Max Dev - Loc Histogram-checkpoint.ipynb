```python
# import modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data into a dataframe
path = '../data/GP_Alumni_List_Cleaned.csv'
df = pd.read_csv(path).copy()
df
```

```python
# histogram

df['StateOrProvince'].hist(figsize=(20,5), bins=50, rwidth=.8)
```

```python
# value count graph
value_counts = df['StateOrProvince'].value_counts()

plt.figure(figsize=(20, 6))
plt.bar(value_counts.index, value_counts.values)
```
